### Handling missing values:

###### drop missing values:

```python

df = pd.DataFrame(np.random.randn(5, 3), index=['a', 'c', 'e', 'f', 'h'],
                     columns=['one', 'two', 'three'])
df
```

```python
df2=df.reindex(['a','b','c','d','e','f','g','h'])
df2
```

```python
df.dropna()
```

```python
pd.isna(df2['one'])
```

```python
pd.notna(df2['one'])
```

```python
df2.fillna('missing')
```

```python
df2['one'].values
```

## Reading different data sources:

```python
from io import StringIO, BytesIO
```

```python
data=('col1,col2,col3\n'
     'x,y,1\n'
     'a,b,2\n'
     'c,d,3')
```

```python
data
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data))
```

#### Reading from specific columns:

```python
df=pd.read_csv(StringIO(data),usecols=lambda x:x.upper() in ['COL1','COL3'])
```

```python
df
```

```python
data = ('a,b,c,d\n'
            '1,2,3,4\n'
            '5,6,7,8\n'
            '9,10,11')
print(data)
```

```python
df=pd.read_csv(StringIO(data),dtype={'b':int,'c':np.float,'a':'Int64'})
```

```python
df
```

```python
df.dtypes
```

### index columns and training delimeters:

```python
data = ('index,a,b,c\n'
           '4,apple,bat,5.7\n'
            '8,orange,cow,10')
```

```python
data
```

```python
pd.read_csv(StringIO(data))
```

```python
pd.read_csv(StringIO(data),index_col=0)
```

```python
data = ('a,b,c\n'
           '4,apple,bat,\n'
            '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data))
```

```python
pd.read_csv(StringIO(data),index_col=False)
```

```python
## Combining usecols and index_col
data = ('a,b,c\n'
        '4,apple,bat\n'
        '8,orange,cow')
```

```python
pd.read_csv(StringIO(data), usecols=['b', 'c'],index_col=False)
```

```python
## Quoting and Escape Characters¶. Very useful in NLP

data = 'a,b\n"hello, \\"Bob\\", nice to see you",5'
```

```python
pd.read_csv(StringIO(data),escapechar='\\')
```
